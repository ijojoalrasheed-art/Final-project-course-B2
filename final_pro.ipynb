```python
import gradio as gr
import joblib
import pandas as pd

model = joblib.load("fuel_model.pkl")


def predict_fuel(passengers, hours, revenue, cost):

    df = pd.DataFrame([{
        "Passengers": passengers,
        "Flight_Hours": hours,
        "Total_Revenue": revenue,
        "Total_Cost": cost
    }])

    df = pd.get_dummies(df)
    df = df.reindex(columns=model.feature_names_in_, fill_value=0)

    pred = model.predict(df)[0]

    if pred == "High":
        return "🔴 HIGH FUEL CONSUMPTION ⛽✈️\n⚠️ Warning: Excess fuel usage detected!"
    else:
        return "🟢 LOW FUEL CONSUMPTION ⛽✈️\n✨ Efficient & Optimized Flight!"



with gr.Blocks(theme=gr.themes.Glass()) as app:

    gr.Markdown(
        """
        # ✈️⛽ AIRLINE FUEL INTELLIGENCE SYSTEM
        ### 🔥 Real-Time Aircraft Fuel Prediction Dashboard
        ---
        ### 📊 Enter flight operational data below
        """
    )


    with gr.Row():
        passengers = gr.Number(label="👥 Number of Passengers", precision=0)

        hours = gr.Number(label="⏱ Flight Hours")

    with gr.Row():
        revenue = gr.Number(label="💰 Total Revenue (USD)")

        cost = gr.Number(label="⛽ Fuel & Operational Cost (USD)")


    btn = gr.Button("🚀 ANALYZE FUEL CONSUMPTION", variant="primary")


    output = gr.Textbox(label="📢 Prediction Result", lines=3)

    btn.click(
        predict_fuel,
        inputs=[passengers, hours, revenue, cost],
        outputs=output
    )


    gr.Markdown(
        """
        ---
        🛫 Powered by AI • Airline Operations Optimization System
        ⛽ Helping airlines reduce fuel waste & improve efficiency
        """
    )

app.launch()
```
